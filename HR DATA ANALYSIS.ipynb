**DATA PRE-PROCESSING**

```python
data.isnull().sum()
```

**CATAGORICAL DATA HANDLING**

```python
data.Attrition = data.Attrition.map({'Yes':1, 'No':8})
```

```python
data.BusinessTravel = data.BusinessTravel.map({"Travel_Frequently':1, 'Travel Rarely': 2, 'fion-Travel":0})
```

```python
data.Department = data.Department.map({'Research & Development': 2, 'Sales": 1, "Human Resources':8})
```

```python
data.EducationField = data.EducationField.map({'Life Sciences': 5, 'Medical':4, 'Marketing':3, 'Technical Degree': 2, 'Other':1, 'Human Resources':0})
```

```python
data.Gender = pd.get_dummies(data.Gender, drop_first = True)
```

```python
data.JobRole = data.JobRole.map({'Laboratory Technician':8, 'Sales Executive':7, 'Research Scientist':6, 'Sales Representative':5, 'Human Resources':4, 'Hanufacturing Director':3, 'Healthcare Representative':2, 'Manager': 1, 'Research & Development':0})
```

```python
data.MaritalStatus = data.MaritalStatus.map({'Single': 2, 'Married': 1, 'Divorced':0})
```

```python
data.OverTime = data.OverTime.map({"Yes":1,"No":0})
```

```python
data.info()
```

####**FEATURE SELECTION**

**DROPING UNIQUE ONES**

```python
data.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis = 1, inplace = True)
```

**MODEL CREATION**

```python
x = data.drop("Attrition",axis = 1)
y = data["Attrition"]
```

**BALANCING THE TARGET COLUMN USING SMOTE**

```python
from collections import Counter
from sklearn.impute import SimpleImputer

from imblearn.over_sampling import SMOTE
sm = SMOTE()
print("unbalanced data: ",Counter(y))

imputer = SimpleImputer(strategy='mean')
x_imputed = imputer.fit_transform(x)
x_sm, y_sm = sm.fit_resample(x_imputed, y)
print("balanced data: ", Counter(y_sm))
```

**TRAIN-TEST-SPLIT**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_sm,y_sm, random_state = 42,test_size = 0.25)
```

```python
data.Attrition.value_counts()
```

**MODEL BUILDING**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

**PREDICTION**

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
```

```python
print(classification_report(y_test,y_pred))
```

**CHECKING DATA IS FULLY TRAINED OR NOT**

```python
train_pred = dt.predict(x_train)
```

```python
print(classification_report(y_train, train_pred))
```

**RANDOM FOREST TECHNIQUE**
